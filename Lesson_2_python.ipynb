```python
shop = [['каретка', 1200], ['шатун', 1000], ['седло', 300], ['педаль', 100], ['седло', 1500],
        ['рама', 12000], ['обод', 2000], ['шатун', 200], ['седло', 2700]]

name = input("Название детали: ")

sum = 0
count = 0
for i in shop:
    if i[0] == name:
        count+=1
        sum+=i[1]

print(f"Количетво деталей - {count}")
print(f"Общая стоимость - {sum}")
```

```python
guests = ['Петя', 'Ваня', 'Саша', 'Лиза', 'Катя']
text = ""

while text != "Пора спать":
    print(f"Сейчас на вечеринке {len(guests)} человек")
    text = input("Гость пришел или ушел? ")
    if text == "Пора спать":
        print("Вечеринка закончилась, все легли спать")
    else:
        name = input("Имя гостя ")
    if text == "пришел" and len(guests) <= 5:
        guests.append(name)
        print(f"Привет, {name}!")
    elif text == "ушел":
        guests.remove(name)
        print(f"Пока, {name}!")
    elif text == "пришел" and len(guests) > 5:
        print(f"Прости, {name}, но мест нет")
```

```python
import random

violator_songs = [
    ['World in My Eyes', 4.86],
    ['Sweetest Perfection', 4.43],
    ['Personal Jesus', 4.56],
    ['Halo', 4.9],
    ['Waiting for the Night', 6.07],
    ['Enjoy the Silence', 4.20],
    ['Policy of Truth', 4.76],
    ['Blue Dress', 4.29],
    ['Clean', 5.83]
]

n = int(input("Сколько песен выбрать? "))

list_of_songs = [random.choice(violator_songs) for x in range(n)]

sum = 0
for index, i in enumerate(list_of_songs):
    print(f"Название {index+1}-й песни: {i[0]}")
    sum+=i[1]

print(f"Общее время звучания песен: {round(sum,2)} минуты")
```

```python
list1 = []
list2 = []

for index, i in enumerate(range(3)):
    i = int(input(f"Введите {index+1}-е число для первого списка "))
    list1.append(i)

for index, i in enumerate(range(7)):
    i = int(input(f"Введите {index+1}-е число для второго списка "))
    list2.append(i)

print(f"Первый список: {list1}")
print(f"Второй список: {list2}")

list1.extend(list2)

print(f"Новый список с уникальными элементами: {sorted(list(set(list1)), reverse=True)}")
```

```python
n = int(input("Количество коньков: "))
k = int(input("Количество людей: "))

list1 = []
list2 = []

for index, i in enumerate(range(n)):
    i = int(input(f"Размер {index+1}-й пары: "))
    list1.append(i)

for index, i in enumerate(range(k)):
    i = int(input(f"Размер ноги {index+1}-го человека: "))
    list2.append(i)

list_com = []
for i in list2:
    for j in list1:
        if i <= j:
            list_com.append(i)

print(f"Наибольшее кол-во людей, которые могут взять ролики: {len(set(list_com))}")
```

```python
n = int(input("Кол-во человек: "))
k = int(input("Какое число в считалке? "))
print("Значит, выбывает каждый", k, "человек.")

l1 = list(range(1, n+1))
vyb = 0

for i in range(n-1):
    print('Текущий круг людей', l1)
    
    start = vyb % len(l1)
    vyb = (start + k - 1) % len(l1)
    
    print('Начало счёта с номера', l1[start])
    print('Выбывает человек под номером', l1[vyb])
    
    l1.remove(l1[vyb])
    
print('Остался человек под номером', l1[0])
```

```python
n = int(input("Количество друзей: "))
k = int(input("Количество расписок: "))

s = dict()
l = []

for i in range(1, k+1):
    s['vzyal'] = int(input("Кому: "))
    s['dal'] = int(input("От кого: "))
    s['money'] = int(input("Сколько: "))
    l.append(s.copy())
    
dict_new = dict()

for j in range(1, n+1):
    sum_vzyal = 0
    sum_dal = 0
    for i in range(len(l)):
        if l[i]['vzyal'] == j:
            sum_vzyal-=l[i]['money']
        elif l[i]['dal'] == j:
            sum_dal+=l[i]['money']
    dict_new[j] = sum_vzyal+sum_dal

print("Баланс друзей: ")
for key, value in dict_new.items():
    print(f"{key}: {value}")
```
